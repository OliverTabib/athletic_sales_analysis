```python
# Import Libraries and Dependencies
import pandas as pd
```

### 1. Combine and Clean the Data
#### Import CSVs

```python
# Read the CSV files into DataFrames.
df2020 = pd.read_csv('Resources/athletic_sales_2020.csv')
df2021 = pd.read_csv('Resources/athletic_sales_2021.csv')

```

#### Combine the sales data by rows.

```python
# Combine the 2020 and 2021 sales DataFrames on the rows and reset the index.
frames = [df2020,df2021]
df = pd.concat(frames)
df.reset_index()
display(df)
```

```python
# Check if any values are null.
null_df = df.isnull()
display(null_df)
```

```python
# Check the data type of each column
df.dtypes
```

```python
# Convert the "invoice_date" to a datetime datatype
df.invoice_date = pd.to_datetime(df.invoice_date)
```

### 2. Determine which Region Sold the Most Products

#### Using `groupby`

```python
# Show the number products sold for region, state, and city.
# Rename the sum to "Total_Products_Sold".
total_products_sold = df.groupby(['region', 'state', 'city']).agg(Total_Products_Sold=('units_sold', 'sum')).sort_values(by='Total_Products_Sold', ascending=False)




# Show the top 5 results.
total_products_sold.head()
                            
```

#### Using `pivot_table`

```python
# Show the number products sold for region, state, and city.
total_products_pivot = df.pivot_table(index=['region', 'state', 'city'], values='units_sold', aggfunc='sum').rename(columns={'units_sold': 'Total_Products_Sold'}).sort_values(by='Total_Products_Sold', ascending=False)
total_products_pivot.head()

# Rename the "units_sold" column to "Total_Products_Sold"


# Show the top 5 results.
```

### 3. Determine which Region had the Most Sales

#### Using `groupby`

```python
# Show the total sales for the products sold for each region, state, and city.
# Rename the "total_sales" column to "Total Sales"
total_sales_grouped = df.groupby(['region', 'state', 'city']).agg(Total_Sales=('total_sales', 'sum')).sort_values(by='Total_Sales', ascending=False)


# Show the top 5 results.
total_sales_grouped.head()
```

#### Using `pivot_table`

```python
# Show the total sales for the products sold for each region, state, and city.
total_sales_pivot = df.pivot_table(index=['region', 'state', 'city'], values='total_sales', aggfunc='sum').rename(columns={'total_sales': 'Total Sales'}).sort_values(by='Total Sales', ascending=False)


# Optional: Rename the "total_sales" column to "Total Sales"


# Show the top 5 results.
total_sales_pivot.head()
```

### 4. Determine which Retailer had the Most Sales

#### Using `groupby`

```python
# Show the total sales for the products sold for each retailer, region, state, and city.
# Rename the "total_sales" column to "Total Sales"
retailer_sales_grouped = df.groupby(['retailer', 'region', 'state', 'city']).agg(Total_Sales=('total_sales', 'sum')).sort_values(by='Total_Sales', ascending=False)

# Show the top 5 results.
retailer_sales_grouped.head()
```

### 5. Determine which Retailer Sold the Most Women's Athletic Footwear

```python
# Filter the sales data to get the women's athletic footwear sales data.
womens_footwear_df = df[df['product'] == "Women's Athletic Footwear"]
```

#### Using `groupby`

```python
# Show the total number of women's athletic footwear sold for each retailer, region, state, and city.
# Rename the "units_sold" column to "Womens_Footwear_Units_Sold"
womens_footwear_sales_grouped = womens_footwear_df.groupby(['retailer', 'region', 'state', 'city']).agg(Womens_Footwear_Units_Sold=('units_sold', 'sum')).sort_values(by='Womens_Footwear_Units_Sold', ascending=False)

# Show the top 5 results.
womens_footwear_sales_grouped.head()
```

#### Using `pivot_table`

```python
# Show the total number of women's athletic footwear sold for each retailer, region, state, and city.
# Rename the "units_sold" column to "Womens_Footwear_Units_Sold"
womens_footwear_sales_pivot = womens_footwear_df.pivot_table(index=['retailer', 'region', 'state', 'city'], values='units_sold', aggfunc='sum').rename(columns={'units_sold': 'Womens_Footwear_Units_Sold'}).sort_values(by='Womens_Footwear_Units_Sold', ascending=False)


# Show the top 5 results.
womens_footwear_sales_pivot.head()
```

### 5. Determine the Day with the Most Women's Athletic Footwear Sales

```python
# Create a pivot table with the 'invoice_date' column is the index, and the "total_sales" as the values.
daily_sales_pivot = womens_footwear_df.pivot_table(index='invoice_date', values='total_sales', aggfunc='sum').rename(columns={'total_sales': 'Total Sales'})



# Optional: Rename the "total_sales" column to "Total Sales"


# Show the table.
display(daily_sales_pivot)
```

```python
# Resample the pivot table into daily bins, and get the total sales for each day.
# Sort the resampled pivot table in ascending order on "Total Sales".
daily_sales_sorted = daily_sales_pivot.resample('D').sum().sort_values(by='Total Sales', ascending=False)


```

### 6.  Determine the Week with the Most Women's Athletic Footwear Sales

```python
# Resample the daily sales pivot table to weekly bins and sum up the sales for each week
weekly_sales = daily_sales_pivot.resample('W').sum()

# Sort the resampled data in ascending order based on "Total Sales"
weekly_sales_sorted = weekly_sales.sort_values(by='Total Sales', ascending=True)

# Display the sorted weekly sales
weekly_sales_sorted
```
